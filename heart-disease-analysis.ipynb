# 6) Find how many people have heart disease, and how many of them have not 

```python
plt.figure(figsize=(6, 5))
sns.countplot(x='target', hue='target', data=data, palette='colorblind', legend=False)

plt.xlabel("Heart Disease Status")
plt.ylabel("Count")
plt.title("Distribution of Heart Disease")

plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])


plt.tight_layout()
plt.show()
```

# 7) Find count of males and females

```python
data['sex'].value_counts()
```

```python
plt.figure(figsize=(8, 6))

sns.violinplot(x='sex', y='thalach', hue='target', data=data, palette='husl', split=True)

plt.xticks([0, 1], ['Female', 'Male'])

plt.title('Max Heart Rate by Gender and Heart Disease')
plt.xlabel('Gender')
plt.ylabel('Max Heart Rate (thalach)')

plt.legend(title='Heart Disease', labels=['No', 'Yes'])

plt.tight_layout()

plt.show()
```

# 8) Find gender distribution according to target variable
